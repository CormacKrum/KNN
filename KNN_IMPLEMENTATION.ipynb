```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
plt.style.use("seaborn")
```

```python
X,y = make_blobs(n_samples=100,random_state=42,centers=2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
query=np.array([[2,10],[1,3],[6,4],[3,6]])
plt.scatter(query[:,0],query[:,1],c='red')
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
def knn(X,y,query,k=5):
    prediction=[]
    for q in range(query.shape[0]):
        vals=[]
        m = X.shape[0]
        for i in range(m):
            d = dist(query[q],X[i])
            vals.append((d,y[i]))
        vals = sorted(vals)
        vals = vals[:k]
        vals = np.array(vals)
        new_vals = np.unique(vals[:,1],return_counts=True)
        max_fre_index = new_vals[1].argmax()
        pred=new_vals[0][max_fre_index]
        prediction.append(pred)
    return prediction
```

```python
knn(X,y,query)
```
